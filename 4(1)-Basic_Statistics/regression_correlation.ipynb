```python
### Predictor Variable과 Response Variable을 설정해주세요!
```

```python
ad_x = ad[['TV', 'radio', 'newspaper']]
ad_y = ad['sales']
```

```python
ad_x.head()
```

```python
ad_y.head()
```

```python
### Multiple Linear Regression을 수행해봅시다!
```

```python
ad_x1 = sm.add_constant(ad_x)
ad_x1.head()
```

```python
model = sm.OLS(ad_y, ad_x1)
result = model.fit()
result.summary()
```
